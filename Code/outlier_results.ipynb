```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```
df = pd.read_csv('after_outlier.csv')
bdf = pd.read_csv('before_outlier.csv')
```

```
bdf.describe()
```

```
df.describe()
```

```
def Distribution2(columne,data,i):
    fig, ax = plt.subplots(1,2, figsize = (15,5))
    font_dict = {'fontsize': 14}
    plt.suptitle(columne, fontsize=16, y=1.05)
    title=['Before Distribution','After Distribution']
    ax = np.ravel(ax)
    if i==1:
        sns.set(style='whitegrid')
        sns.kdeplot(data=data,x=columne ,ax = ax[0],color='r').set_title(title[i])
        sns.boxplot(data=data,x=columne ,ax = ax[1],palette='magma').set_title(title[i])
    else:
        sns.set(style='whitegrid')
        sns.kdeplot(data=data,x=columne ,ax = ax[0],color='#2171b5').set_title(title[i])
        sns.boxplot(data=data,x=columne ,ax = ax[1],color='#2171b5').set_title(title[i])
        
    ax = np.reshape(ax, (1, 2))
    plt.tight_layout()
```

```
Distribution2(columne='Age',data=bdf,i=0)
Distribution2(columne='Age',data=df,i=1)
```

```
Distribution2(columne='Num_Credit_Inquiries',data=bdf,i=0)
Distribution2(columne='Num_Credit_Inquiries',data=df,i=1)
```

```
Distribution2(columne='Num_Credit_Card',data=bdf,i=0)
Distribution2(columne='Num_Credit_Card',data=df,i=1)
```

```
Distribution2(columne='Num_of_Loan',data=bdf,i=0)
Distribution2(columne='Num_of_Loan',data=df,i=1)
```

```
Distribution2(columne='Num_Bank_Accounts',data=bdf,i=0)
Distribution2(columne='Num_Bank_Accounts',data=df,i=1)
```

```
Distribution2(columne='Num_of_Delayed_Payment',data=bdf,i=0)
Distribution2(columne='Num_of_Delayed_Payment',data=df,i=1)
```

```
Distribution2(columne='Interest_Rate',data=bdf,i=0)
Distribution2(columne='Interest_Rate',data=df,i=1)
```
